## 1. Data 준비

### Train 이미지 사이즈 변경
28X28 사이즈로 변경

````python
from PIL import Image
import os, glob
````

````python
# 사이즈 변경 함수
def resize_images(img_path):
	images=glob.glob(img_path + "/*.jpg")
    
	print(len(images), " images to be resized.")


	target_size=(28,28) # 파일마다 28X28 사이즈로 바꾸어 저장
	for img in images:
		old_img=Image.open(img)
		new_img=old_img.resize(target_size,Image.ANTIALIAS)
		new_img.save(img, "JPEG")
    
	print(len(images), " images resized.")
````

````python
# train_set의 가위,바위,보 이미지의 사이즈 변경

image_dir_path = os.getenv("HOME") + "/aiffel/rock_scissor_paper/scissor"
resize_images(image_dir_path)
print("가위 이미지 resized")

image_dir_path = os.getenv("HOME") + "/aiffel/rock_scissor_paper/rock"
resize_images(image_dir_path)
print("바위 이미지 resized")

image_dir_path = os.getenv("HOME") + "/aiffel/rock_scissor_paper/paper"
resize_images(image_dir_path)
print("보 이미지 resized")
````

각각 라벨링을 하고

`load_data()` 라는 데이터를 불러오는 함수 작성

````python
import numpy as np

def load_data(img_path, number_of_data=7175):  # 가위바위보 이미지 개수 총합에 주의하세요.
    # 가위 : 0, 바위 : 1, 보 : 2
    img_size=28
    color=3
    #이미지 데이터와 라벨(가위 : 0, 바위 : 1, 보 : 2) 데이터를 담을 행렬(matrix) 영역을 생성합니다.
    imgs=np.zeros(number_of_data*img_size*img_size*color,dtype=np.int32).reshape(number_of_data,img_size,img_size,color)
    labels=np.zeros(number_of_data,dtype=np.int32)

    idx=0
    for file in glob.iglob(img_path+'/scissor/*.jpg'):
        img = np.array(Image.open(file),dtype=np.int32)
        imgs[idx,:,:,:]=img    # 데이터 영역에 이미지 행렬을 복사
        labels[idx]=0   # 가위 : 0
        idx=idx+1

    for file in glob.iglob(img_path+'/rock/*.jpg'):
        img = np.array(Image.open(file),dtype=np.int32)
        imgs[idx,:,:,:]=img    # 데이터 영역에 이미지 행렬을 복사
        labels[idx]=1   # 바위 : 1
        idx=idx+1  
    
    for file in glob.iglob(img_path+'/paper/*.jpg'):
        img = np.array(Image.open(file),dtype=np.int32)
        imgs[idx,:,:,:]=img    # 데이터 영역에 이미지 행렬을 복사
        labels[idx]=2   # 보 : 2
        idx=idx+1
        
    print("학습데이터(x_train)의 이미지 개수는", idx,"입니다.")
    return imgs, labels
````

train_set을 불러오고, train/255.0을 하여 입력을 0~1 사이의 값으로 정규화시켜준다.

````python
image_dir_path = os.getenv("HOME") + "/aiffel/rock_scissor_paper"
(x_train, y_train)=load_data(image_dir_path)
x_train_norm = x_train/255.0   # 입력은 0~1 사이의 값으로 정규화

print("x_train shape: {}".format(x_train.shape))
print("y_train shape: {}".format(y_train.shape))
````

마지막 3은 **RGB채널** 수 를 뜻한다.

## 2. Deeplearning Network 설계

````python
# LeNet 네트워크
import tensorflow as tf
from tensorflow import keras
import numpy as np

# model을 직접 만들어 보세요.
# Hint! model의 입력/출력부에 특히 유의해 주세요. 가위바위보 데이터셋은 MNIST 데이터셋과 어떤 점이 달라졌나요?
model=keras.models.Sequential()
# C1 Layer : 28X28사이즈 이미지들을, 16개의 3X3 필터와 Convolution연산을 해준다.
model.add(keras.layers.Conv2D(16, (3,3), activation='relu', input_shape=(28,28,3)))

# S2 Layer : 얻은 16장의 특성 맵에 대해 subsampling, 2X2필터를 stride 2로 설정하므로 13X13사이즈의 특성맵으로 축소
model.add(keras.layers.MaxPool2D(2,2))

# C3 Layer : 16장의 13X13 특성맵에 컨볼루션하여 32장의 11X11특성맵 산출
model.add(keras.layers.Conv2D(32, (3,3), activation='relu'))

# S4 Layer : 32장의 11X11 특성맵에 대하여 subsampling하여 5X5 특성맵으로 축소
model.add(keras.layers.MaxPooling2D((2,2)))

# C5 Layer : 5X5X32 사이즈의 필터와 컨볼루션 하여 800개의 1X1 특성맵 산출
model.add(keras.layers.Flatten())

# F6 Layer : 32개의 유닛을 가진 FeedForward 신경망, C5결과를 32개의 유닛에 연결
model.add(keras.layers.Dense(32, activation='relu'))

 # output Layer : 3개의 유닛, 최종적으로 이미지가 속한 클래스를 알려준다.
model.add(keras.layers.Dense(3, activation='softmax'))
print('Model에 추가된 Layer 개수: ', len(model.layers))


model.summary()
````

## Deeplearning Network 학습

````python
model.compile(optimizer='adam',
             loss='sparse_categorical_crossentropy',
             metrics=['accuracy'])

model.fit(x_train_norm, y_train, epochs=10)
````

model이 추론해 낸 숫자와 실제 라벨이 다른 경우는 어떤경우?

100개 정도를 이미지로 출력하여 그 원인을 생각해 보겠다.

````python
import matplotlib.pyplot as plt


predicted_result = model.predict(x_test_norm)  # model이 추론한 확률값. 
predicted_labels = np.argmax(predicted_result, axis=1)

idx=0  #1번째 x_test를 살펴보자. 
print('model.predict() 결과 : ', predicted_result[idx])
print('model이 추론한 가장 가능성이 높은 결과 : ', predicted_labels[idx])
print('실제 데이터의 라벨 : ', y_test[idx])


import random
wrong_predict_list=[]
for i, _ in enumerate(predicted_labels):
    # i번째 test_labels과 y_test이 다른 경우만 모아 봅시다. 
    if predicted_labels[i] != y_test[i]:
        wrong_predict_list.append(i)

# wrong_predict_list 에서 랜덤하게 5개만 뽑아봅시다.
samples = random.choices(population=wrong_predict_list, k=100)

for n in samples:
    print("예측확률분포: " + str(predicted_result[n]))
    print("라벨: " + str(y_test[n]) + ", 예측결과: " + str(predicted_labels[n]))
    plt.imshow(x_test[n], cmap=plt.cm.binary)
    plt.show()
````

예측확률분포의 숫자들의 뒤에 붙은 e- 는 지수표현식인데
### 지수표현식
e-05 이면 소수점을 다섯칸 왼쪽으로 옮긴다.

EX) 6.0299026e-05 = 0.000060299026 이다. 고로 e- 뒤에 붙은 숫자가 작을 수록 전체적인 숫자는 큰 숫자가 된다.

### 예측 확률 분포 해석
EX) [6.0299026e-05 9.9993467e-01 5.0438039e-06] = [0, 1, 2]

각각 라벨의 0번째, 1번째, 2번째가 정답일것이다라고 예측한 확률을 나타내는 숫자인데 여기서는 99%의 확률로 1번 레이블이라고 확신을 한것이다.

즉, 이 라벨에서는 99%확률로 1번 라벨(바위)이라고 예측을한것이다. 하지만 정답은 0번 라벨인 가위여서 틀렸다.

### 그 외에 잘못 예측 한 이유 분석
잘못 예측한 이미지 사진을 100개정도 추출하여 본 결과, 

- case 1. 정답은 scissor인데 뒤의 배경이 깔끔하지 않아 rock으로 잘못예측한것과

- case 2. 배경은 깔끔한데 정답은 rock이지만 paper로 잘못 예측 한 케이스로 나뉜다.


case1의 경우 배경이 28X28사이즈로 resize로 했을 때 이것저것 섞여서 잘못 예측이 가능하다고 충분히 생각이 되는데,

case2 같은 경우에는 배경이 깔끔한데 불구하고 왜 잘못 예측을 한지는 잘 모르겠다.
