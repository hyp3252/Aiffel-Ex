# 얼굴임베딩
## FaceNet
- 네트워크 뒤에 L2 Normalization을 거쳐 임베딩을 만들어 내고 여기에 Triplet Loss를 사용
### L2 Normalization
- 모델결과물의 L2 Distance를 구한 후에 Normalization을 해주는 과정
- Triplet Loss : 세 개의 데이터 쌍을 이용해 계산하는 손실함수로 네트워크를 학습시킬 수 있다.
- A,C는 같은 사람의 얼굴 사진에서 나온 임베딩 벡터, B는 다른사람의 얼굴 사진에서 만들어진 임베딩 벡터가 되도록 데이터셋을 구성
- ``` Face Recognition``` 에는 ``` FaceNet``` 얼굴 임베딩 모델이 탑재되어 있다.

### 이미지에서 얼굴 임베딩 추출

````python
# file path 가져오기
import os
dir_path= os.getenv("HOME")+'/aiffel/face_embedding/images/'
file_list = os.listdir(dir_path)
print("file_list:{}".format(file_list))

# image 가져오기
image_file = os.path.join(dir_path, 'obama.jpg')
face = get_cropped_face(image_file)

# 얼굴 영역을 가지고 얼굴 임베딩벡터를 구하는 함수
def get_face_embedding(face):
    return face_recognition.face_encodings(face)


embedding = get_face_embedding(face)
print(embedding)
````

````python
# images 디렉토리의 모든 이미지 파일의 임베딩을 구하여 dict구조에 담아 리턴하는 함수
def get_face_embedding_dict(dir_path):
    file_list = os.listdir(dir_path)
    embedding_dict={}
    
    for file in file_list:
        img_path = os.path.join(dir_path, file)
        face = get_cropped_face(img_path)
        embedding = get_face_embedding(face)
        if len(embedding) > 0: # face가 제대로 검출되지 않으면 len(embedding)==0 인경우가 발생
            embedding_dict[os.path.splitext(file)[0]] = embedding[0] # 파일이름.확장자 명을 분리 시킨후에 파일이름만 담기.
            
    return embedding_dict
````

# 얼굴 임베딩 공간의 시각화
- 임베딩 벡터는 넘파이배열로 되어있어, 감이 잘 오지 않는다.
- 고차원 데이터를 저차원으로 바꾼 임베딩처럼 다차원 벡터를 시각화 하기 위하여 차원 축소를 하는 방법들이 있다.
- 예 : **PCA, T-SNE** 등
- Tensorflow의 ```Projector```는 고차원 벡터를 차원 축소 기법을 사용하여 시각화해준다.

## PCA
- 주성분 분석이라는 방법.
- 모든 차원의 축에 따른 값의 변화도인 분산을 확인하고 그 중 변화가 가장 큰 주요한 축을 남기는 방법

## T-SNE
- 고차원 상에서 먼 거리를 저차원 상에서도 멀리 배치되도록 차원축소하는 방식
- random하게 목표하는 차원에 데이터들을 배치한 후 각 데이터들을 고차원상에서의 배치와 비교를 하면서 위치를 변경해 주는 알고리즘
