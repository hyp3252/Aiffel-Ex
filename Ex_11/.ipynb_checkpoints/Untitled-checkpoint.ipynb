### 캐글 Chest X-Ray images

- train, test, val 로 구성
- 5,856개의 X-Ray이미지와 2개의 범주가있다.(폐렴/정상)

````python
# import package
import os, re
import random, math
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings(action='ignore')
````

````python
AUTOTUNE = tf.data.experimental.AUTOTUNE

# X-ray 이미지 사이즈
IMAGE_SIZE = [180,180]
# 데이터 경로
ROOT_PATH = os.path.join(os.getenv('HOME'),'aiffel')
TRAIN_PATH = ROOT_PATH + '/chest_xray/data/train/*/*'
VAL_PATH = ROOT_PATH + '/chest_xray/data/val/*/*'
TEST_PATH = ROOT_PATH + '/chest_xray/data/test/*/*'

# 하이퍼파라미터
BATCH_SIZE = 16 # 미니 배치
EPOCHS = 25

print(ROOT_PATH)
````

### 데이터불러오기

````python
train_filenames = tf.io.gfile.glob(TRAIN_PATH)
test_filenames = tf.io.gfile.glob(TEST_PATH)
val_filenames = tf.io.gfile.glob(VAL_PATH)

print(len(train_filenames))
print(len(test_filenames))
print(len(val_filenames))
````

train 89%, test 10.7%, val 0.3% 의 비율로 val개수가 너무 없으므로 train+val 하여 8:2로 다시 나눈다

````python
# train_set + val_set
filenames = tf.io.gfile.glob(TRAIN_PATH)
filenames.extend(tf.io.gfile.glob(VAL_PATH))

# train:val = 8:2 로
train_size = math.floor(len(filenames)*0.8) # 정수형으로 나와야하므로 반내림 처리
random.seed(8)
random.shuffle(filenames)
train_filenames = filenames[:train_size]
val_filenames = filenames[train_size:]

print(len(train_filenames))
print(len(val_filenames))
````

````python
# 정상 및 폐렴 이미지 수
print(f'Normal image path\n{filenames[0]}') # 정상부분
print(f'Pneumonia image path\n{filenames[2000]}') # 폐렴부분
````

````python
COUNT_NORMAL = len([filename for filename in train_filenames if "NORMAL" in filename])
print('Trainset Normal images : {}'.format(COUNT_NORMAL))

COUNT_PNEUMONIA = len([filename for filename in train_filenames if "PNEUMONIA" in filename])
print('Trainset Pneumonia images : {}'.format(COUNT_PNEUMONIA))
````

- 결과를 보면 정상 : 폐렴 = 1:3 의 비율로 있다.
- 클래스 불균형은 차후에 조정
### minibatch는 왜 사용하는가?
- 딥러닝에서 한번의 iteration을 위해 들어가는 input data는 보통 batch라고 하여 데이터를 한그룹으로 사용한다.
- mini-batch는 input data로 한개를 쓰는경우와 전체데이터셋을 쓰는 경우의 장점을 모두 얻기 위함이다.
**forward+backpropa+update= 1 iteration**

#### 데이터 한개 쓰는경우
- 장점: iteration 한번 수행하는데 소요되는 시간이 짧다.
- 단점: 데이터 전체의 경향반영이 어렵다.
- 단점: 하드웨어적으로 비효율적.

#### 전체데이터를 쓰는 경우
- 장점: 전체데이터로 한걸음 내딛는다. cost function값을 줄이는 양질의 이동을 한다.
- 단점: 데이터셋의 크기가 커지면 iteration시간이 매우길다. learning rate로도 어렵다.
- 단점: 하드웨어입장에서 부담

### 결론
- 적당히 빠르고 적당히 정확한 길을 찾기 위하여 mini-batch 사용

````python
train_list_ds = tf.data.Dataset.from_tensor_slices(train_filenames)
val_list_ds = tf.data.Dataset.from_tensor_slices(val_filenames)
````

````python
# train set, val set 개수확인
TRAIN_IMG_COUNT = tf.data.experimental.cardinality(train_list_ds).numpy()
print('Training images: {}'.format(TRAIN_IMG_COUNT))

VAL_IMG_COUNT = tf.data.experimental.cardinality(val_list_ds).numpy()
print('Validation images: {}'.format(VAL_IMG_COUNT))
````

- 라벨데이터가 따로없으므로 경로에 'NORMAL'이나 'PNEUMONIA'를 이용하여 라벨데이터를 만든다.

````python
# 경로의 [-2]번째에 양/음성 판단가능
def get_label(file_path):
    parts = tf.strings.split(file_path, os.path.sep)
    return parts[-2]=="PNEUMONIA"
````

- 이미지데이터는 현실적으로 사이즈가 제각각일 가능성이 높다.
- 이미지 사이즈를 통일시킨다.
- 함수생성

````python
# image 변환함수
def decode_img(img):
    img = tf.image.decode_jpeg(img, channels=3) # uint8 tensor로 변환
    img = tf.image.convert_image_dtype(img, tf.float32) # float32 로 변환
    img = tf.image.resize(img, IMAGE_SIZE) # image resizing
    return img


# image 경로 입력 시 이미지와 라벨을 읽어온다.
def process_path(file_path):
    label = get_label(file_path) # 라벨 검출 함수
    img = tf.io.read_file(file_path) # 이미지 읽어오기
    img = decode_img(img) # image 변환함수
    return img,label
````

````python
# train set, val set을 만든다.
train_ds = train_list_ds.map(process_path, num_parallel_calls=AUTOTUNE) # num_parallel_calls?
val_ds = val_list_ds.map(process_path, num_parallel_calls=AUTOTUNE)
````

````python
# image , label 확인
for image,label in train_ds.take(1):
    print('image shape:{}'.format(image.numpy().shape))
    print('label: {}'.format(label.numpy()))
````

#### test set에도 똑같이 만든다.

````python
test_list_ds = tf.data.Dataset.list_files(TEST_PATH)
TEST_IMAGE_COUNT = tf.data.experimental.cardinality(test_list_ds).numpy()
test_ds = test_list_ds.map(process_path, num_parallel_calls = AUTOTUNE)
test_ds = test_ds.batch(BATCH_SIZE)
print(TEST_IMAGE_COUNT)
print(test_ds)
````

- Tensorflow에서는 tf.data 파이프라인을 사용해서 학습데이터를 효율적으로 사용할 수 있도록 한다.

- prepare_for_training()함수를 정의하여 데이터를 변환

- shuffle을 사용하며 고정 크기 버퍼를 유지하고 해당 버퍼에서 무작위로 균일하게 다음 요소를 선택

- repeat()을 사용하면 epoch를 진행하면서 여러 번 데이터셋을 불러오는데, 여러번 데이터셋을 사용할 수 있게 해준다.

EX) 100개의 데이터를 10번 반복 시 1000개의 데이터가 필요. 자동으로 데이터를 맞춰준다.

- batch()를 사용하면 BATCH_SIZE에서 정한만큼 배치로 주어짐

EX) 100개의 데이터를 10개의 batch로 나누면 각 batch에 10개의 데이터

- prefetch()를 사용하면 학습데이터를 나눠서 읽어오기 때문에, GPU에서 첫번째 데이터 학습하는 동안, 두번째 데이터를 준비할 수 있어서 유휴상태를 줄인다.

````python
def prepare_for_training(ds, shuffle_buffer_size=1000): # shuffle_buffer_size : 전체 데이터에서 파티션을 나눠서 각 파티션을 random shuffle
    ds = ds.shuffle(buffer_size = shuffle_buffer_size)
    ds = ds.repeat()
    ds = ds.batch(BATCH_SIZE)
    ds = ds.prefetch(buffer_size=AUTOTUNE)
    return ds

val_ds = prepare_for_training(val_ds)
````

### 데이터 시각화
- train에 있는 batch 중 첫번째를 추출,
- 추출 된 배치를 image와 label 데이터셋으로 나눈다.

````python
def show_batch(image_batch, label_batch):
    plt.figure(figsize=(10,10))
    for n in range(BATCH_SIZE):
        ax = plt.subplot(5, math.ceil(BATCH_SIZE/5), n+1) # math.ceil : EX) 5.0001 --> 6으로 표현
        plt.imshow(image_batch[n])
        if label_batch[n]:
            plt.title('PNEUMONIA')
        else:
            plt.title('NORMAL')
        plt.axis('off')
        
image_batch, label_batch = next(iter(train_ds))
show_batch(image_batch.numpy(), label_batch.numpy())
````

### CNN 모델링
- Convolution을 두번진행하고 Batch normalization으로 Gradient vanishing, Gradient Exploding을 해결

````python
def conv_block(filters):
    block = tf.keras.Sequential([
        tf.keras.layers.SeparableConv2D(filters,3,activation='relu',padding='same'),
        tf.keras.layers.SeparableConv2D(filters,3,activation='relu',padding='same'),
        tf.keras.layers.BatchNormalization(),
        tf.keras.layers.MaxPool2D()
    ])
    return block
````

````python
# Dense block
def dense_block(units, dropout_rate):
    block = tf.keras.Sequential([
        tf.keras.layers.Dense(units, activation='relu'),
        tf.keras.layers.BatchNormalization(),
        tf.keras.layers.Dropout(dropout_rate)
    ])
    
    
    return block
````

### 현재만든 CNN모델이 기존과는 다른점

- Batch Normalization과 Dropout 두 가지가 동시에 사용된다.
- 일반적으로는 잘 사용하지 않거나 금기시된다.
- 의견은 갈리지만 성능 향상에 도움을 준다고도 되어있다.

````python
# model
def build_model():
    model=tf.keras.Sequential([
        tf.keras.Input(shape=(IMAGE_SIZE[0], IMAGE_SIZE[1], 3 )),
        tf.keras.layers.Conv2D(16,3,activation='relu', padding='same'),
        tf.keras.layers.Conv2D(16,3,activation='relu', padding='same'),
        tf.keras.layers.MaxPool2D(),
        
        conv_block(32),
        conv_block(64),
        conv_block(128),
        tf.keras.layers.Dropout(0.2),
        conv_block(256),
        tf.keras.layers.Dropout(0.2),
        tf.keras.layers.Flatten(),
        
        dense_block(512,0.7), # 512 unit, dropout(0.7)
        dense_block(128,0.5),
        dense_block(64, 0.3),
        tf.keras.layers.Dense(1, activation='sigmoid')
    ])
    
    return model
````

### Data imbalance 처리
- 폐렴 음성과 양성의 비율이 밸런스가 맞지않아 학습효과가 좋지 않을 수 있다.
- ``` Weight balancing```사용 : training set의 각 데이터에서 loss 계산할때 특정 클래스의 데이터에 loss값을 더 갖도록 가중치 부여


### 불균형 처리 중 다른 방식(Over sampling & under sampling)
#### Oversampling
- 적은 class의 복사본을 만들어서, majority class와 숫자를 맞춘다.

#### Under sampling
- Majority class의 일부만을 선택하고, minority class 데이터는 최대한 많이 사용한다.
- 이 경우에는 majority class가 원본데이터와 비교하여 대표성이 있어야한다.

````python
weight_for_0 = (1/COUNT_NORMAL)*(TRAIN_IMG_COUNT)/2.0 # normal 이미지에 사용할 weight
weight_for_1 = (1/COUNT_PNEUMONIA)*(TRAIN_IMG_COUNT)/2.0 # Pneumonia 이미지 weight

class_weight = {0 : weight_for_0, 1:weight_for_1}

print('Weight for Normal:{0:.2f}\nWeight for Pneumonia:{1:.2f}'.format(weight_for_0, weight_for_1))
````

### model training
- GPU사용, build_model()을 model에 선언
- 이미지 라벨은 2개이므로 binary cross entropy
- 성과측정 metrics == accuracy, precision, recall

````python
with tf.device('/GPU:0'):
    model = build_model()
    
    METRICS = ['accuracy', tf.keras.metrics.Precision(name='precision'), tf.keras.metrics.Recall(name='recall')]
    
    model.compile(optimizer='adam', loss='binary_crossentropy',metrics=METRICS)
````

### model fit

````python
with tf.device('/GPU:0'):
    history = model.fit(
        train_ds,
        steps_per_epoch = TRAIN_IMG_COUNT//BATCH_SIZE,
        epochs = EPOCHS,
        validation_data = val_ds,
        validation_steps = VAL_IMG_COUNT//BATCH_SIZE,
        class_weight = class_weight,
    )
````

````python
fig, ax = plt.subplots(1, 4, figsize=(20, 3))
ax = ax.ravel()

for i, met in enumerate(['precision', 'recall', 'accuracy', 'loss']):
    ax[i].plot(history.history[met])
    ax[i].plot(history.history['val_' + met])
    ax[i].set_title('Model {}'.format(met))
    ax[i].set_xlabel('epochs')
    ax[i].set_ylabel(met)
    ax[i].legend(['train', 'val'])
````

````python
loss, accuracy, precision, recall = model.evaluate(test_ds)
print(f'Loss: {loss},\nAccuracy: {accuracy},\nPrecision: {precision},\nRecall: {recall}')
````
