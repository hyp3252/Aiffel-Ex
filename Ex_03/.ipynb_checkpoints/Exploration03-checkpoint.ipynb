```python
# module import
import cv2
import matplotlib.pyplot as plt
import numpy as np
print('import complete')
```

```python
#   
import os
my_image_path = os.getenv("HOME")+"/aiffel/camera_sticker/images/ILSVRC2012_test_00065880.png"
my_image_path2 = os.getenv("HOME")+"/aiffel/camera_sticker/images/ILSVRC2012_test_00097330.png"
my_image_path3 = os.getenv("HOME")+"/aiffel/camera_sticker/images/messi.png"
my_image_path4 = os.getenv("HOME")+"/aiffel/camera_sticker/images/person.png"
# image resize
    
img_bgr = cv2.imread(my_image_path)
img_bgr = cv2.resize(img_bgr, (1200,1200))

img_bgr2 = cv2.imread(my_image_path2)
img_bgr2 = cv2.resize(img_bgr2, (1200,1200))

img_bgr3 = cv2.imread(my_image_path3)
img_bgr3 = cv2.resize(img_bgr3, (1200,1200))

img_bgr4 = cv2.imread(my_image_path4)
img_bgr4 = cv2.resize(img_bgr4, (1200,1200))


#    
img_show = img_bgr.copy()
img_show_bgr = img_bgr.copy()
img_show2 = img_bgr2.copy()
img_show_bgr2 = img_bgr2.copy()
img_show3 = img_bgr3.copy()
img_show_bgr3 = img_bgr3.copy()
img_show4 = img_bgr4.copy()
img_show_bgr4 = img_bgr4.copy()

#### 2 2  ##
fig = plt.figure()
rows = 2
cols = 2


# 11   
ax1 = fig.add_subplot(rows, cols, 1)
ax1.imshow(cv2.cvtColor(img_show, cv2.COLOR_BGR2RGB))
ax1.set_title('1')
ax1.axis("off")


# 12   
ax2 = fig.add_subplot(rows, cols,2)
ax2.imshow(cv2.cvtColor(img_show2, cv2.COLOR_BGR2RGB))
ax2.set_title('2')
ax2.axis("off")

# 21   

ax3 = fig.add_subplot(rows, cols,3)
ax3.imshow(cv2.cvtColor(img_show3, cv2.COLOR_BGR2RGB))
ax3.set_title('3')
ax3.axis("off")

# 22   
ax4 = fig.add_subplot(rows, cols,4)
ax4.imshow(cv2.cvtColor(img_show4, cv2.COLOR_BGR2RGB))
ax4.set_title('4')
ax4.axis("off")

plt.show()
```

```python
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
plt.show()

img_rgb2 = cv2.cvtColor(img_bgr2, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb2)
plt.show()

img_rgb3 = cv2.cvtColor(img_bgr3, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb3)
plt.show()

img_rgb4 = cv2.cvtColor(img_bgr4, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb4)
plt.show()
```

```python
# hog detector
import dlib
detector_hog = dlib.get_frontal_face_detector()
print('done')
```

```python
# bounding box
dlib_rects = detector_hog(img_rgb, 1)
dlib_rects2 = detector_hog(img_rgb2, 1)
dlib_rects3 = detector_hog(img_rgb3, 1)
dlib_rects4 = detector_hog(img_rgb4, 1)
```

```python
#  
print("1 boundingbox:{0}\n2 boundingbox:{1}\n3 boundingbox:{2}\n4 boundingbox:{3}"
      .format(dlib_rects, dlib_rects2, dlib_rects3, dlib_rects4))

for dlib_rect in dlib_rects:
    l = dlib_rect.left()
    t = dlib_rect.top()
    r = dlib_rect.right()
    b = dlib_rect.bottom()
    cv2.rectangle(img_show,(l,t),(r,b),(0,255,0),2 , lineType=cv2.LINE_AA)
    
for dlib_rect2 in dlib_rects2:
    l = dlib_rect2.left()
    t = dlib_rect2.top()
    r = dlib_rect2.right()
    b = dlib_rect2.bottom()
    cv2.rectangle(img_show2,(l,t),(r,b),(0,255,0),2 , lineType=cv2.LINE_AA)
    
for dlib_rect3 in dlib_rects3:
    l = dlib_rect3.left()
    t = dlib_rect3.top()
    r = dlib_rect3.right()
    b = dlib_rect3.bottom()
    cv2.rectangle(img_show3,(l,t),(r,b),(0,255,0),2 , lineType=cv2.LINE_AA)
    
for dlib_rect4 in dlib_rects4:
    l = dlib_rect4.left()
    t = dlib_rect4.top()
    r = dlib_rect4.right()
    b = dlib_rect4.bottom()
    cv2.rectangle(img_show4,(l,t),(r,b),(0,255,0),2 , lineType=cv2.LINE_AA)
```

```python
img_show_rgb = cv2.cvtColor(img_show, cv2.COLOR_BGR2RGB)
plt.imshow(img_show_rgb)
plt.show()

img_show_rgb2 = cv2.cvtColor(img_show2, cv2.COLOR_BGR2RGB)
plt.imshow(img_show_rgb2)
plt.show()

img_show_rgb3 = cv2.cvtColor(img_show3, cv2.COLOR_BGR2RGB)
plt.imshow(img_show_rgb3)
plt.show()

img_show_rgb4 = cv2.cvtColor(img_show4, cv2.COLOR_BGR2RGB)
plt.imshow(img_show_rgb4)
plt.show()
```

```python
#  landmark  .
import os
model_path = os.getenv("HOME") + "/aiffel/camera_sticker/models/shape_predictor_68_face_landmarks.dat"
# RGB dblib.rectangle  dlib.full_object_detection
landmark_predictor = dlib.shape_predictor(model_path)
print("done")
```

```python
list_landmarks = []
list_landmarks2 = []
list_landmarks3 = []
list_landmarks4 = []

for dlib_rect in dlib_rects:
    points = landmark_predictor(img_rgb, dlib_rect)
    list_points = list(map(lambda p:(p.x, p.y), points.parts()))
    list_landmarks.append(list_points)
    
for dlib_rect2 in dlib_rects2:
    points = landmark_predictor(img_rgb2, dlib_rect2)
    list_points = list(map(lambda p:(p.x, p.y), points.parts()))
    list_landmarks2.append(list_points)
    
for dlib_rect3 in dlib_rects3:
    points = landmark_predictor(img_rgb3, dlib_rect3)
    list_points = list(map(lambda p:(p.x, p.y), points.parts()))
    list_landmarks3.append(list_points)
    
for dlib_rect4 in dlib_rects4:
    points = landmark_predictor(img_rgb4, dlib_rect4)
    list_points = list(map(lambda p:(p.x, p.y), points.parts()))
    list_landmarks4.append(list_points)
    
print(len(list_landmarks[0]))
print(len(list_landmarks2[0]))
print(len(list_landmarks3[0]))
print(len(list_landmarks4[0]))
# points   parts()     . 
```

```python
#  
for landmark in list_landmarks:
    for point in landmark:
        cv2.circle(img_show, point, 2, (0,255,255), -1)
        
for landmark2 in list_landmarks2:
    for point in landmark2:
        cv2.circle(img_show2, point, 2, (0,255,255), -1)
        
        
for landmark3 in list_landmarks3:
    for point in landmark3:
        cv2.circle(img_show3, point, 2, (0,255,255), -1)
        
        
for landmark4 in list_landmarks4:
    for point in landmark4:
        cv2.circle(img_show4, point, 2, (0,255,255), -1)


img_show_rgb = cv2.cvtColor(img_show, cv2.COLOR_BGR2RGB)
img_show_rgb2 = cv2.cvtColor(img_show2, cv2.COLOR_BGR2RGB)
img_show_rgb3 = cv2.cvtColor(img_show3, cv2.COLOR_BGR2RGB)
img_show_rgb4 = cv2.cvtColor(img_show4, cv2.COLOR_BGR2RGB)

fig = plt.figure()
rows = 2
cols = 2


# 11   
ax1 = fig.add_subplot(rows, cols, 1)
ax1.imshow(img_show_rgb)
ax1.set_title('1')
ax1.axis("off")


# 12   
ax2 = fig.add_subplot(rows, cols,2)
ax2.imshow(img_show_rgb2)
ax2.set_title('2')
ax2.axis("off")

# 21   

ax3 = fig.add_subplot(rows, cols,3)
ax3.imshow(img_show_rgb3)
ax3.set_title('3')
ax3.axis("off")

# 22   
ax4 = fig.add_subplot(rows, cols,4)
ax4.imshow(img_show_rgb4)
ax4.set_title('4')
ax4.axis("off")

plt.show()
```

```python
import os
sticker_path = os.getenv('HOME')+'/aiffel/camera_sticker/images/cat-whiskers.png'
img_sticker = cv2.imread(sticker_path)
img_sticker = cv2.resize(img_sticker, (w,h))

sticker_path = os.getenv('HOME')+'/aiffel/camera_sticker/images/cat-whiskers.png'
img_sticker2 = cv2.imread(sticker_path)
img_sticker2 = cv2.resize(img_sticker2, (w2,h2))

sticker_path = os.getenv('HOME')+'/aiffel/camera_sticker/images/cat-whiskers.png'
img_sticker3 = cv2.imread(sticker_path)
img_sticker3 = cv2.resize(img_sticker3, (w3,h3))

sticker_path = os.getenv('HOME')+'/aiffel/camera_sticker/images/cat-whiskers.png'
img_sticker4 = cv2.imread(sticker_path)
img_sticker4 = cv2.resize(img_sticker4, (w4,h4))

print (img_sticker.shape)
print (img_sticker2.shape)
print (img_sticker3.shape)
print (img_sticker4.shape)
```

```python
refined_x = x-w//2
refined_y = y-h//2
print ('(x,y) : (%d,%d)'%(refined_x, refined_y))

refined_x2 = x2-w2//2
refined_y2 = y2-h2//2
print ('(x2,y2) : (%d,%d)'%(refined_x2, refined_y2))

refined_x3 = x3-w3//2
refined_y3 = y3-h3//2
print ('(x3,y3) : (%d,%d)'%(refined_x3, refined_y3))

refined_x4 = x4-w4//2
refined_y4 = y4-h4//2
print ('(x4,y4) : (%d,%d)'%(refined_x4, refined_y4))
```

```python
# 
sticker_area = img_bgr[refined_y:refined_y + img_sticker.shape[0], refined_x:refined_x+img_sticker.shape[1]]
img_bgr[refined_y:refined_y+img_sticker.shape[0], refined_x:refined_x+img_sticker.shape[1]] = \
    np.where(img_sticker==255,sticker_area,img_sticker).astype(np.uint8)

sticker_area2 = img_bgr2[refined_y2:refined_y2 + img_sticker2.shape[0], refined_x2:refined_x2+img_sticker2.shape[1]]
img_bgr2[refined_y2:refined_y2+img_sticker2.shape[0], refined_x2:refined_x2+img_sticker2.shape[1]] = \
    np.where(img_sticker2==255,sticker_area2,img_sticker2).astype(np.uint8)

sticker_area3 = img_bgr3[refined_y3:refined_y3 + img_sticker3.shape[0], refined_x3:refined_x3+img_sticker3.shape[1]]
img_bgr3[refined_y3:refined_y3+img_sticker3.shape[0], refined_x3:refined_x3+img_sticker3.shape[1]] = \
    np.where(img_sticker3==255,sticker_area3,img_sticker3).astype(np.uint8)

sticker_area4 = img_bgr4[refined_y4:refined_y4 + img_sticker4.shape[0], refined_x4:refined_x4+img_sticker4.shape[1]]
img_bgr4[refined_y4:refined_y4+img_sticker4.shape[0], refined_x4:refined_x4+img_sticker4.shape[1]] = \
    np.where(img_sticker4==255,sticker_area4,img_sticker4).astype(np.uint8)






plt.imshow(cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB))
plt.show()
plt.imshow(cv2.cvtColor(img_bgr2, cv2.COLOR_BGR2RGB))
plt.show()
plt.imshow(cv2.cvtColor(img_bgr3, cv2.COLOR_BGR2RGB))
plt.show()
plt.imshow(cv2.cvtColor(img_bgr4, cv2.COLOR_BGR2RGB))
plt.show()
```
