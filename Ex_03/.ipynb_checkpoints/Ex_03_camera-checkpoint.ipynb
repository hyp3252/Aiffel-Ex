````python
# module import
import cv2
import matplotlib.pyplot as plt
import numpy as np
print("import complete")
````

````python
# 준비한 이미지 읽어들이기
import os
my_image_path = os.getenv("HOME")+"/aiffel/camera_sticker/images/image.png"
# cv로 이미지 읽고 640X360크기로 resize
img_bgr = cv2.imread(my_image_path)
img_bgr = cv2.resize(img_bgr, (640,640))

# 출력용 이미지 따로 보관
img_show = img_bgr.copy()


plt.imshow(img_bgr)
plt.show()
````

사진에 푸른빛이 감도는 이유는 opencv특징이다
## opencv
- 예외적으로 BGR 을 사용
- 붉은색과 푸른색이 바뀌어 출력된다

````python
# BGR -> RGB로 변환
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
plt.show()
````

## 얼굴 검출
- dlib의 face detector는 Histogram of Oriented Gradients(HOG)를 사용하여 SVM의 sliding window로 얼굴을 찾는다.

- sliding window : 사각형으로 표시되는 윈도우가 전체 이미지를 왼쪽에서 오른쪽으로, 위 아래로 이동하면서 생성하는 모든영역을 모델의 입력값으로 넣어주어 결과를 얻는 방식

````python
# hog detector
import dlib
detector_hog = dlib.get_frontal_face_detector()
print('done')
````

````python
# bounding box 추출
# dlib은 RGB이미지를 입력으로 받으므로 변환
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)

# 두번째 파라미터는 이미지 피라미드 수
dlib_rects = detector_hog(img_rgb, 1)
print('done')
````

- upsampling : 데이터의 크기를 키우는 것.

````python
# 화면에 출력
print(dlib_rects)

for dlib_rect in dlib_rects:
    l = dlib_rect.left()
    t = dlib_rect.top()
    r = dlib_rect.right()
    b = dlib_rect.bottom()
    
    cv2.rectangle(img_show,(l,t),(r,b),(0,255,0),2 , lineType=cv2.LINE_AA)
img_show_rgb = cv2.cvtColor(img_show, cv2.COLOR_BGR2RGB)
plt.imshow(img_show_rgb)
plt.show()
````

## face landmark loacalization
스티커를 섬세하게 적용하기 위해서는 이목구비의 위치를 아는 것이 중요

detection의 결과물인 bounding box로 잘라낸(crop) 얼굴 이미지를 이용

### Object keypoint estimation 알고리즘
객체 내부의 점을 찾는 기술

1) top-down : bounding box를 찾고 box내부의 keypoint를 예측

2) bottom-up : 이미지 전체의 keypoint를 먼저 찾고 point 관계를 이용해 군집화해서 box생성

예제에서는 1번을 다룬다.

## dlib landmark localization
dlib의 제공되는 모델 사용.

weight file다운로드

````python
# 저장한 landmark 모델을 불러온다.
import os
model_path = os.getenv("HOME") + "/aiffel/camera_sticker/models/shape_predictor_68_face_landmarks.dat"
# RGB이미지와 dblib.rectangle을 입력받아 dlib.full_object_detection반환
landmark_predictor = dlib.shape_predictor(model_path)
print("done")
````

````python
list_landmarks = []
for dlib_rect in dlib_rects:
    points = landmark_predictor(img_rgb, dlib_rect)
    list_points = list(map(lambda p:(p.x, p.y), points.parts()))
    list_landmarks.append(list_points)
    
print(len(list_landmarks[0]))
# points는 객체이기 때문에 parts() 함수로 개별 위치에 접근 가능. 
````

````python
# 랜드마크 출력
for landmark in list_landmarks:
    for point in landmark:
        cv2.circle(img_show, point, 2, (0,255,255), -1)
        
        
img_show_rgb = cv2.cvtColor(img_show, cv2.COLOR_BGR2RGB)
plt.imshow(img_show_rgb)
plt.show()
````

````python
# 좌표확인
for dlib_rect, landmark in zip(dlib_rects, list_landmarks):
    print(landmark[30])
    x = landmark[30][0]
    y = landmark[30][1] - dlib_rect.height()//2
    w = dlib_rect.width()
    h = dlib_rect.height()
    print('(x,y) : (%d, %d)'%(x,y))
    print('(w,h) : (%d, %d)'%(w,h))
````

````python
# 스티커 이미지 읽고 적용
import os
sticker_path = os.getenv("HOME") + "/aiffel/camera_sticker/images/king.png"
img_sticker = cv2.imread(sticker_path)
img_sticker = cv2.resize(img_sticker, (w,h))
print(img_sticker.shape)
````

````python
refined_x = x-w//2 # left
refined_y = y-h # top
print('(x,y):(%d,%d)'%(refined_x, refined_y))
````

````python
# 음수에 대한 예외처리
img_sticker = img_sticker[-refined_y:]
print(img_sticker.shape)
````

````python
refined_y=0
print('(x,y) : (%d,%d)'%(refined_x,refined_y))
````

````python
# 원본이미지에 스티커
sticker_area = img_bgr[refined_y:img_sticker.shape[0], refined_x:refined_x+img_sticker.shape[1]]
img_bgr[refined_y:img_sticker.shape[0], refined_x:refined_x+img_sticker.shape[1]] = \
    np.where(img_sticker==0,sticker_area,img_sticker).astype(np.uint8)
plt.imshow(cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB))
plt.show()
````
